```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame({
    'Sepal length':iris.data[:,0],
    'Sepal width':iris.data[:,1],
    'Petal length':iris.data[:,2],
    'Petal width':iris.data[:,3],
    'class':iris.target
})
```

```python
iris_data
# iris dataset link: https://github.com/dotnet/machinelearning/blob/main/test/data/iris.txt
# class 0 = Iris-setosa
# class 1 = Iris-versicolor
# calss 2 = Iris-virginica
```

```python
fig = plt.figure(figsize=(5,5))
plt.scatter(iris_data['Sepal length'],iris_data['Sepal width'],c=iris.target)
```

```python
train_y = iris_data.pop('class')
train_y = np.array(train_y)
train_x = iris_data
train_x = np.array(train_x)
```

```python
knn = KNeighborsClassifier()
knn.fit(train_x,train_y)
```

```python
l = train_x.mean(axis=0)
name = ['Sepal length ','Sepal width  ','Petal length ','Petal width  ']
for i in range(4):
    num = float(input(name[i]+': '))
    if(num!=0):
        l[i] = num

arr = np.array(l).reshape(1,-1)
knn.predict(arr)
```

```python
y_test = knn.predict(arr)
#plt.gcf().set_size_inches(8, 8)
class_colors = ['r', 'g', 'b']
train_colors = [class_colors[c] for c in train_y]
test_colors = [class_colors[c] for c in y_test]
plt.scatter(train_x[:, 0], train_x[:, 1],color=train_colors, marker='o', s=32, alpha=0.2)
plt.scatter(arr[:, 0], arr[:, 1],color=test_colors, marker='*', s=128.0)
plt.title('Sepal')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```

```python
plt.scatter(train_x[:, 2], train_x[:, 3],color=train_colors, marker='o', s=32, alpha=0.2)
plt.scatter(arr[:, 2], arr[:, 3],color=test_colors, marker='*', s=128.0)
plt.title('Petal')
plt.xlabel('Length')
plt.ylabel('Width')
plt.show()
```
